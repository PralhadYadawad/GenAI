```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# Load pre-trained model and tokenizer
model_name = 'gpt2-medium'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Ensure the model is in evaluation mode
model.eval()

def generate_tweet(topic, mood):
    """Generate Tweet text."""
    mood_prompt = f"{mood} " if mood else ""
    prompt = f"{mood_prompt}Tweet about {topic}: "

    # Encode the input
    inputs = tokenizer.encode(prompt, return_tensors='pt')
    attention_mask = torch.ones(inputs.shape, device=inputs.device)

    # Generate the response
    outputs = model.generate(
        inputs,
        attention_mask=attention_mask,
        max_length=50,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        top_p=0.95,
        top_k=50,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode the generated text
    tweet = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Remove the prompt from the generated text
    tweet = tweet.replace(prompt, "").strip()

    return tweet
```

```python
topic = input("Enter the topic: ")
mood = input("Enter the mood (optional): ")
tweet = generate_tweet(topic, mood)
print(f"\nGenerated Tweet:\n{tweet}")
```

```python
!pip install transformers torch

import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
model_name = 'gpt2-medium'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Ensure the model is in evaluation mode
model.eval()

def calculate_perplexity(model, tokenizer, text):
    """
    Calculate perplexity for a given text using the GPT-2 model.

    Parameters:
    model (transformers.GPT2LMHeadModel): The GPT-2 model.
    tokenizer (transformers.GPT2Tokenizer): The tokenizer for the GPT-2 model.
    text (str): The text for which perplexity is to be calculated.

    Returns:
    float: The perplexity score for the input text.
    """
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt")
    input_ids = inputs["input_ids"]

    # Get model outputs (logits) and calculate loss
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
        loss = outputs.loss  # Cross-entropy loss

    # Perplexity is the exponential of the loss
    perplexity = torch.exp(loss)
    return perplexity.item()

def generate_tweet(topic, mood=None):
    """
    Generate a tweet based on the given topic and mood.

    Parameters:
    topic (str): The topic for the tweet.
    mood (str, optional): The mood for the tweet. If not provided, defaults to an empty string.

    Returns:
    str: The generated tweet.
    """
    # Default mood to an empty string if not provided
    if mood is None:
        mood = ""

    mood_prompt = f"{mood} " if mood else ""
    prompt = f"{mood_prompt}Tweet about {topic}: "

    # Encode the input text
    inputs = tokenizer.encode(prompt, return_tensors='pt')

    # Generate the tweet response
    outputs = model.generate(
        inputs,
        max_length=50,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        top_p=0.95,
        top_k=50,
        pad_token_id=tokenizer.eos_token_id,
        temperature=1.0,  # Optional: to introduce some variability in generation
    )

    # Decode the generated text
    tweet = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Remove the prompt from the generated text
    tweet = tweet.replace(prompt, "").strip()

    return tweet

# Take user input for topic and mood

```

```python
topic = input("Enter the topic: ")
mood = input("Enter the mood (optional): ")

# Generate tweet based on inputs
tweet = generate_tweet(topic, mood)

# Print the generated tweet
print(f"\nGenerated Tweet:\n{tweet}")

# Calculate and print Perplexity for the generated tweet
ppl = calculate_perplexity(model, tokenizer, tweet)
print(f"\nPerplexity of the generated tweet: {ppl}")
```
